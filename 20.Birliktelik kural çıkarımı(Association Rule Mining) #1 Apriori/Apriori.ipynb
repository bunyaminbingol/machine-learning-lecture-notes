```python
# Kütüphaneleri import ediyoruz

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
veriler = pd.read_csv("veriler/sepet.csv", header= None)

t = []
for i in range(0,7501):
    t.append([str(veriler.values[i,j]) for j in range(0,20)]) #burada her bir satırı alıyorum i ile kullanılan for döngüsü ile sonra j for döngüsü ile satır içindeki elemanları alıyorum.
veriler
```

```python
t # burada liste içerisinde liste oluşturduk çünkü apriori algoritması bu şekilde çalışıyor, aşağıda görseli var.
# her bir satırda 19 elemanlı bir liste var ve bu listenin içinde alınan ürünler var.
```

```python
from apyori import apriori 
kurallar = apriori(t, min_support = 0.01, min_confidance = 0.2, min_lift = 3, min_length = 2) # length değeri ile kaç tane eleman istediğimizi giriyoruz yani en az 2 li ürünleri getirsin ikili kampanya yapmak istiyoruz.
list(kurallar)
```
