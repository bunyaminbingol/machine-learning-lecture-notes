```python
#Ktphaneleri aktif ediyoruz
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

# verimizi ekelim

veriler = pd.read_csv("veriler/musteriler.csv")
veriler
```

```python
# verileri oluturalm

X = veriler.iloc[:,3:].values
```

```python
from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(n_clusters=3, affinity="euclidean", linkage="ward") # "n_clusters" ile ka cluster istiyorsak onu belirliyoruz, "affinity" dorusallktan bahsediyor aslnda burada mesafeler arasndaki farka bakacak, "linkage" ise kmeler arasndaki fark hangi method ile bulacamz belirtiyoruz.
Y_tahmin = ac.fit_predict(X)
Y_tahmin # her bir data point iin girdii cluster' belirtecek.
```

```python
# bunlar izelim,

plt.scatter(X[Y_tahmin==0,0],X[Y_tahmin==0,1], c="red")
plt.scatter(X[Y_tahmin==1,0],X[Y_tahmin==1,1], c="blue")
plt.scatter(X[Y_tahmin==2,0],X[Y_tahmin==2,1], c="black")
```

```python
# dendrogram' izelim,

import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method="ward"))
```
