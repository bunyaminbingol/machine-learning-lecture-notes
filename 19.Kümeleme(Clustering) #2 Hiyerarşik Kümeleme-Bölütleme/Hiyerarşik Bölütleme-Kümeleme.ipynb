
* **burada görüldüğü gibi "Agglomerative" algoritması ile ilk baştaki k-means algoritmasının farkını keskin bir şekilde görebiliyoruz.**

```python
#Kütüphaneleri aktif ediyoruz
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

# verimizi çekelim

veriler = pd.read_csv("veriler/musteriler.csv")
veriler
```

```python
# verileri oluşturalım

X = veriler.iloc[:,3:].values
```

```python
from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(n_clusters=3, affinity="euclidean", linkage="ward") # "n_clusters" ile kaç cluster istiyorsak onu belirliyoruz, "affinity" doğrusallıktan bahsediyor aslında burada mesafeler arasındaki farka bakacak, "linkage" ise kümeler arasındaki farkı hangi method ile bulacağımızı belirtiyoruz.
Y_tahmin = ac.fit_predict(X)
Y_tahmin # her bir data point için girdiği cluster'ı belirtecek.
```

```python
# bunları çizelim,

plt.scatter(X[Y_tahmin==0,0],X[Y_tahmin==0,1], c="red")
plt.scatter(X[Y_tahmin==1,0],X[Y_tahmin==1,1], c="blue")
plt.scatter(X[Y_tahmin==2,0],X[Y_tahmin==2,1], c="black")
```

```python
# dendrogram'ı çizelim,

import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method="ward"))
```
