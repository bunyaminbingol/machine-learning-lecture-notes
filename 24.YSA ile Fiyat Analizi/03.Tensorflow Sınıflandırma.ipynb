# Sınıflandırma Modeli Oluşturma

```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 15)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential  
from tensorflow.keras.layers import Dense, Activation, Dropout 
from tensorflow.keras.callbacks import EarlyStopping
```

**Dropout ve EarlyStopping kütüphanelerini overfitting yani ezberleme olmaması için kullanacağız.**

```python
x_train.shape
```

```python
model = Sequential()

# Aşağıda oluşturacağım model önceden oluşturulmuş bir kurala bağlı, böyle olmak zorunda değil.
# Bir sinir ağı oluşturduğumuz zaman, başlangıç olarak özellik sayımız kaç ise nöronumuzu da ona göre aynı sayı yapmaya çalışırız.
model.add(Dense(units = 30, activation= "relu"))
model.add(Dense(units = 15, activation= "relu")) # Bu iki ara katman için 1 ile 30 arasında bir değer koyarız
model.add(Dense(units = 15, activation= "relu")) # Bu iki ara katman için 1 ile 30 arasında bir değer koyarız
model.add(Dense(units = 1, activation= "relu")) # Sınıflandırma modellerinde son gizli katman 1 olur ve bu katmanın activation fonction'ı sigmoid olmalı çünkü sigmoid 0 ve 1 arasında bir değer verir 

model.compile(loss = "binary_crossentropy", optimizer = "adam")  # Sınıflandırmada loss için binary_crossentropy kullanabiliriz.
```

```python
# Epochs değerini abartalım ve sonucu görelim:

model.fit(x = x_train, y = y_train, epochs = 700, validation_data = (x_test, y_test), verbose = 1)
```

```python
model.history.history
```

```python
modelKaybi = pd.DataFrame(model.history.history)
modelKaybi.plot()
```

* **Görüldüğü gibi overfitting oluyor**
* **Tüm veriyi tek tek sınıflanfırmaya çalışıyor ve bu sınıflandırma sonucu ezberlemiş oluyor, bunun olmasını istemiyoruz. bunun için şunu yapabiliyoruz; mesela val_loss'u izleyip şunu diyebiliyoruz; eğer val_loss saçma bir şekilde artarsa o zaman epoch'u durdur, tekrar train etme diyoruz. Bunun için EarlyStopping kütüphanesini kullanıyoruz.**

# Erken Durdurma

```python
# Yine aynı modeli alıyoruz:

model = Sequential()

model.add(Dense(units = 30, activation= "relu"))
model.add(Dense(units = 15, activation= "relu")) 
model.add(Dense(units = 15, activation= "relu")) 
model.add(Dense(units = 1, activation= "relu")) 

model.compile(loss = "binary_crossentropy", optimizer = "adam")  
```

```python
earlyStopping = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=25)
# monitor: Hangisini gözlemleyeceğiz onu söyle diyor.

# patience: belirlediğimiz epoch miktarından sonra modele bakıyor ve model de bir iyileşme yoksa modeli durduruyor.
# Bunu az verirsek erken, çok verirsek geç durmaya neden olabilir.

# verbose: ne kadar yazdıracağı

# mode: "{" auto "," min "," max "}" seçeneklerinden biri. "Min" modunda, val_lossu en azda tutmaya çalıştığımız 
# manasına geliyor. başka birşey tutmaya çalıştığımızda mesela accuracy, "max"'ı tercih edebilirdik.
```

```python
# epoch sayısını değiştirmiyoruz çünkü model buna kendisi karar verecek.
# diğer model.fit'den farklı olarak buraya callbacks ile oluşturduğumuz earlyStopping'i ekliyoruz.
model.fit(x = x_train, y = y_train, epochs = 700, validation_data = (x_test, y_test), verbose = 1, callbacks = [earlyStopping])
```

**Biz burada epoch sayımızı 143 yapmamız lazımmış overfitting olmaması için ve biz bunu bilmiyorduk, öğrendik.**

```python
modelKaybi = pd.DataFrame(model.history.history)
modelKaybi.plot()
```

* **Gördüğümüz gibi gerçekten de baya bir iyileşme var**
* **Bu durumdan halen memnun değilsek bir yöntem daha vari Dropout**

# Dropout

**rate: % kaçsa o kadar layer'ı bizim modelimizden random bir şekilde atmaya başlayacak. 
Böylece eğer çok fazla layer varsa bunları atarak biz daha iyi bir model'e ve sonuca ulaşmış olacağız.
Overfitting'i daha iyi engelleyerek verimli bir sonuç elde etmiş olacağız.**

```python
# Yine aynı modeli alıyoruz ve her katman arasına Dropout koyuyoruz.

model = Sequential()

model.add(Dense(units = 30, activation= "relu"))
model.add(Dropout(rate = 0.5))

model.add(Dense(units = 15, activation= "relu")) 
model.add(Dropout(rate = 0.5))

model.add(Dense(units = 15, activation= "relu")) 
model.add(Dropout(rate = 0.5))

model.add(Dense(units = 1, activation= "relu")) 

model.compile(loss = "binary_crossentropy", optimizer = "adam")  
```

```python
model.fit(x = x_train, y = y_train, epochs = 700, validation_data = (x_test, y_test), verbose = 1, callbacks = [earlyStopping])
```

```python
kayipDf = pd.DataFrame(model.history.history)
kayipDf.plot()
```

# Tahmin Edelim

```python
tahminlerimiz = model.predict_classes(x_test)   # predict_classes ile sınıfları tahmin et diyebiliyoruz.
tahminlerimiz
```

```python
# Tahminlerimizi güzel değerlendirebilmek için birkaç metrik kullanacağız:
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, tahminlerimiz))
```

* **precision: ne kadar duyarlı bir tahmin yaptığımızı gösteriyor**

```python
print(confusion_matrix(y_test,tahminlerimiz))
```

```python
# Başarı Oranı
print(accuracy_score(y_test, tahminlerimiz))
```
