```python
#1.kutuphaneler

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


#veri yukleme

veriler = pd.read_csv('veriler/veriler.csv')
x = veriler.iloc[:,1:4].values # Bağımsız değişkenler, boy, kilo ve yaş
y = veriler.iloc[:,4:].values # bağmlı değişken, cinsiyet
```

```python
veriler
```

```python
# verilerin eğitim ve test için bölünmesi

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 0)
```

```python
x_train
```

```python
# verilerin ölçeklenmesi

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(x_train) 
X_test = sc.transform(x_test) 
```

```python
from sklearn.svm import SVC

svc =  SVC(kernel = 'linear') # Burada doğrusal fonksiyonu kullanmaya karar verdik, verimize göre değiştirebiliriz.
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```
