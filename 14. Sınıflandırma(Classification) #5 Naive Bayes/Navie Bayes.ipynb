**Naive Bayes iin  temel yntemden bahsedeceiz, Gaussian Naive Bayes, Multinomial Naive Bayes ve Bernoulli Naive Bayes bunlarn aslnda zellii, 3'de farkl dalmlar zerinde alyor.**
* Tahmin edeceimiz snf srekli bir deer ise reel saylar olabiliyorsa, ondalkl saylar olabiliyorsa Gaussian Navie Bayes'i kullanmak
> 
* Birbirinden farkl veriler zeride mesela diyelim ki "Erkek" ve "Kadn"dan oluan bir veri yani iki taneden birisi veya arabanzn markas gibi nominal deerler. Bu deerlere numara(1,2,3....) veriyorsak ve bunlarn tahmini iin kullanacaksak o zaman Multinomial Naive Bayes'i kullanabiliriz.
> 
* "1" veya "0" gibi binomial yani nominal olacak ama iki nominalden birisi "kadn - erkek" gibi "yapyor - yapmyor" gibi durumlar varsa o zaman da Bernoulli Navie Bayes'i kullanabiliriz.
> 

```python
#1.kutuphaneler

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


#veri yukleme

veriler = pd.read_csv('veriler/veriler.csv')
x = veriler.iloc[:,1:4].values # Bamsz deikenler, boy, kilo ve ya
y = veriler.iloc[:,4:].values # baml deiken, cinsiyet
```

```python
# verilerin eitim ve test iin blnmesi

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 0)
```

```python
# verilerin leklenmesi

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(x_train) 
X_test = sc.transform(x_test) 
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(X_train, y_train)

y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Baar Oran
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
# Sonu : 0.739961759082218
print(accuracy)
```
