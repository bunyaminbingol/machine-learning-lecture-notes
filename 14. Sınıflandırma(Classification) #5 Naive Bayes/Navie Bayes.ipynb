**Naive Bayes için üç temel yöntemden bahsedeceğiz, Gaussian Naive Bayes, Multinomial Naive Bayes ve Bernoulli Naive Bayes bunların aslında özelliği, 3'de farklı dağılımlar üzerinde çalışıyor.**
* Tahmin edeceğimiz sınıf sürekli bir değer ise reel sayılar olabiliyorsa, ondalıklı sayılar olabiliyorsa Gaussian Navie Bayes'i kullanmak
> 
* Birbirinden farklı veriler üzeride mesela diyelim ki "Erkek" ve "Kadın"dan oluşan bir veri yani iki taneden birisi veya arabanızın markası gibi nominal değerler. Bu değerlere numara(1,2,3....) veriyorsak ve bunların tahmini için kullanacaksak o zaman Multinomial Naive Bayes'i kullanabiliriz.
> 
* "1" veya "0" gibi binomial yani nominal olacak ama iki nominalden birisi "kadın - erkek" gibi "yapıyor - yapmıyor" gibi durumlar varsa o zaman da Bernoulli Navie Bayes'i kullanabiliriz.
> 

```python
#1.kutuphaneler

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


#veri yukleme

veriler = pd.read_csv('veriler/veriler.csv')
x = veriler.iloc[:,1:4].values # Bağımsız değişkenler, boy, kilo ve yaş
y = veriler.iloc[:,4:].values # bağmlı değişken, cinsiyet
```

```python
# verilerin eğitim ve test için bölünmesi

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 0)
```

```python
# verilerin ölçeklenmesi

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(x_train) 
X_test = sc.transform(x_test) 
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(X_train, y_train)

y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Başarı Oranı
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
# Sonuç : 0.739961759082218
print(accuracy)
```
