```python
#1.kutuphaneler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#2.veri onisleme
#2.1.veri yukleme
veriler = pd.read_excel('veriler/Iris.xls')
#pd.read_csv("veriler.csv")
veriler
```

```python
x = veriler.iloc[:,1:4].values #bamsz deikenler
y = veriler.iloc[:,4:].values #baml deiken
```

```python
#verilerin egitim ve test icin bolunmesi
from sklearn.model_selection import train_test_split

x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=0)
```

```python
#verilerin olceklenmesi
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression(random_state=0)
logr.fit(X_train,y_train)

y_pred = logr.predict(X_test)
print(y_pred)
print(y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
# Baar Oran
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
# Sonu : 0.739961759082218
print(accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1, metric='minkowski')
knn.fit(X_train,y_train)

y_pred = knn.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
# Baar Oran
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
# Sonu : 0.739961759082218
print(accuracy)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='rbf')
svc.fit(X_train,y_train)

y_pred = svc.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
print('SVC')
print(cm)
```

```python
# Baar Oran
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
# Sonu : 0.739961759082218
print(accuracy)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)

y_pred = gnb.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
print('GNB')
print(cm)
```

```python
# Baar Oran
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
# Sonu : 0.739961759082218
print(accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion = 'entropy')

dtc.fit(X_train,y_train)
y_pred = dtc.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
print('DTC')
print(cm)
```

```python
# Baar Oran
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
# Sonu : 0.739961759082218
print(accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=10, criterion = 'entropy')
rfc.fit(X_train,y_train)

y_pred = rfc.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print('RFC')
print(cm)
```

```python
# Baar Oran
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
# Sonu : 0.739961759082218
print(accuracy)
```

```python
# 7. ROC , TPR, FPR deerleri 

y_proba = rfc.predict_proba(X_test) # predict_proba X_test'in tahmin olaslklarn verir. 
print(y_test)
print(y_proba)
```

```python
from sklearn import metrics

fpr , tpr , thold = metrics.roc_curve(y_test,y_proba[:,0],pos_label='Iris-virginica')

print(fpr) # False Pozitif Rate
print(tpr) # True Pozitif Rate
print(thold) # Threshold
```
